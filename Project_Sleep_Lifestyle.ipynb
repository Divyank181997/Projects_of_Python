# Impact of Lifestyle on Sleep Health¶

Sleep plays a vital role in maintaining overall health and well-being. However, various lifestyle factors
can significantly impact sleep quality and duration. Understanding the relationship between lifestyle choices
and sleep health is essential for individuals seeking to improve their sleep patterns. 

# Key Features

1. Person ID: An identifier for each individual in the dataset.
2. Gender: The gender of the person (Male/Female).
3. Occupation: The occupation or profession of the person.
4. Sleep Duration (hours): The number of hours the person sleeps per day.
5. Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10
6. Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.
7. Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
8. BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).
9.Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
10. Heart Rate (bpm): The resting heart rate of the person in beats per minute.
11. Daily Steps: The number of steps the person takes per day.
12. Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

# Exploratory Analysis and Visualization

# 1. People from which gender has better quality of sleep?

```python
df1 = df.groupby('Gender')[['Quality of Sleep','Stress Level']].mean()
df1 = df1.reset_index()
df1
```

```python
#PLotting the Graph
plt.figure(figsize = (10,5))
plt.plot(df1.loc[:,'Gender'],df1.loc[:,'Quality of Sleep'],color = 'Blue',label='Quality of Sleep')
plt.plot(df1.loc[:,'Gender'],df1.loc[:,'Stress Level'],color = 'Red',label='Stress Level')
plt.xlabel("Gender")
plt.grid(linewidth=0.5)
plt.legend()
```

From the above data, it can be seen that : 
1. Females have better quality of sleep as compared to males.
2. The cause of it is the lower levels of stress in females as compared to males.

# 2. People from which occupation has better quality of sleep?

```python
df2 = df.groupby('Occupation')[['Quality of Sleep','Stress Level']].mean()
df2 = df2.reset_index()
df2
```

```python
#PLotting the Graph
plt.figure(figsize = (18,5))
plt.plot(df2.loc[:,'Occupation'],df2.loc[:,'Quality of Sleep'],color = 'Blue',label='Quality of Sleep')
plt.plot(df2.loc[:,'Occupation'],df2.loc[:,'Stress Level'],color = 'Red',label='Stress Level')
plt.xlabel("Gender")
plt.grid(linewidth=0.5)
plt.legend()
```

1. Above analysis clearly shows that Sales Representative has the least quality of sleep which can be seen
   from its high levels of stress value that can be seen in the data.
2. On the other hand, Engineer has the best quality of sleep with the least stress levels as compared to other 
   occupations.

# 3. Quality of sleep based on physical activity

```python
df3 = pd.DataFrame(df.groupby('BMI Category')['Quality of Sleep'].mean())
df3 = df3.reset_index()
df3
```

```python
#PLotting the Graph
plt.figure(figsize = (10,5))
plt.plot(df3.loc[:,'BMI Category'],df3.loc[:,'Quality of Sleep'],color = 'Red',label='Quality of Sleep')
plt.xlabel("BMI Category")
plt.ylabel("Quality of Sleep")
plt.grid(linewidth=0.5)
plt.legend()
```

It can be seen from the above graph that healthier an individual is, better is the quality of sleep is he/she having.
Obese and overweight people have the least quality of sleep.

# 4. How does your BMI Category affects your heart rate?

```python
df4 =pd.DataFrame(df.groupby('BMI Category')['Heart Rate'].mean())
df4 = df4.reset_index()
df4
```

```python
fig, ax= plt.subplots(figsize=(6,5))
sns.pointplot(y= 'BMI Category', x= 'Heart Rate', data=df4, ax=ax)
plt.show()
```

It can be seen that a person whose BMI Category is obese is suffering from the highest heart rate which is not healthy
for a person and therefore, he/she should work upon it.

# 5. Relation between Occupation and BMI Category

```python
df5 =pd.DataFrame(df.groupby('Occupation')['BMI Category'].value_counts())
df5
```

From the above data, it can be seen that most of the Nurses are overweight while most of the doctors and engineers are normal

# 6. Relation between Stress level and Sleep Disorder

```python
df7 = pd.DataFrame(df.groupby('Sleep Disorder')['Stress Level'].mean())
df7 = df7.reset_index()
df7
```

```python
c = ['red', 'yellow', 'brown']
plt.bar(df7['Sleep Disorder'],df7['Stress Level'],color=c,width = 0.5)
plt.xlabel("Sleep Disorder")
plt.ylabel("Stress Level")
plt.title("Relation between Stress level and Sleep Disorder")
plt.show()
```

From the above data, it is easy to conclude that people with the least stress levels are the ones who are
not suffering from any kind of sleep disorder and are considered healthy.
