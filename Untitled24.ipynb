# Description of dataset

This comprehensive dataset provides a wealth of information about all countries worldwide, covering a wide range of indicators and attributes. It encompasses demographic statistics, economic indicators, environmental factors, healthcare metrics, education statistics, and much more. With every country represented, this dataset offers a complete global perspective on various aspects of nations, enabling in-depth analyses and cross-country comparisons.

# Key Features

1. Country: Name of the country.
2. Density (P/Km2): Population density measured in persons per square kilometer.
3. Agricultural Land (%): Percentage of land area used for agricultural purposes.
4. Land Area (Km2): Total land area of the country in square kilometers.
5. CO2 Emissions: Carbon dioxide emissions in tons.
6. Forested Area (%): Percentage of land area covered by forests.
7. Gross Primary Education Enrollment (%): Gross enrollment ratio for primary education.
8. Gross Tertiary Education Enrollment (%): Gross enrollment ratio for tertiary education.
9. Population: Total population of the country.
10. Population: Labor Force Participation (%): Percentage of the population that is part of the labor force.
11. Unemployment Rate: Percentage of the labor force that is unemployed.
12. Urban Population: Percentage of the population living in urban areas.


# Data Preparation and Cleaning

```python
df.isnull().sum()
```

```python
# Drop the columns that are not in use
df1 = df.drop(columns=['Abbreviation','Armed Forces size','Birth Rate','Calling Code','Capital/Major City',
                      'CPI','CPI Change (%)','Currency-Code','Fertility Rate','Gasoline Price','GDP','Infant mortality',
                      'Largest city','Life expectancy','Maternal mortality ratio','Official language','Out of pocket health expenditure',
                      'Physicians per thousand','Tax revenue (%)','Total tax rate','Latitude','Longitude','Minimum wage','Land Area(Km2)'])
df1
```

```python
#Remove the null values as none of the 'NaN' values in this case can be taken as 0.
df1 = df1.dropna()
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
# As we can see from above that many of the columns should be of float type for the analysis to be done.
```

```python
# Replacing all the commas and % signs so that it is easy to convert the data type of columns
df1['Agricultural Land( %)'] = df1['Agricultural Land( %)'].str.replace("%","")
df1['Forested Area (%)'] = df1['Forested Area (%)'].str.replace("%","")
df1['Co2-Emissions'] = df1['Co2-Emissions'].str.replace(",","")
df1['Gross tertiary education enrollment (%)'] = df1['Gross tertiary education enrollment (%)'].str.replace("%","")
df1['Gross primary education enrollment (%)'] = df1['Gross primary education enrollment (%)'].str.replace("%","")
df1['Population'] = df1['Population'].str.replace(",","")
df1['Density\n(P/Km2)'] = df1['Density\n(P/Km2)'].str.replace(",","")
df1['Population: Labor force participation (%)'] = df1['Population: Labor force participation (%)'].str.replace("%","")
df1['Unemployment rate'] = df1['Unemployment rate'].str.replace("%","")
df1['Urban_population'] = df1['Urban_population'].str.replace(",","")
df1
```

```python
# Changing the data type of all the columns into whatever data type that is helpful for analysis
df1 = df1.astype({'Agricultural Land( %)':'float','Co2-Emissions':'float','Gross primary education enrollment (%)':'float',
                 'Gross tertiary education enrollment (%)':'float','Population':'float',
                 'Population: Labor force participation (%)':'float','Unemployment rate':'float',
                 'Urban_population':'float','Forested Area (%)':'float','Density\n(P/Km2)':'float',})
```

# 1. Study labor market dynamics through indicators such as labor force participation and unemployment rates.

```python
df2 = df1.groupby('Country')['Unemployment rate','Population: Labor force participation (%)'].sum()
df2
```

10 Countries with highest Unemployment rate are : -

```python
unempl = df2.sort_values(by = ['Unemployment rate']).tail(5)
unempl = unempl.reset_index()
unempl
```

```python
#Top 5 Countries with highest population density
c = ['red', 'yellow', 'brown', 'blue', 'orange']
plt.bar(unempl['Country'],unempl['Unemployment rate'],color=c,width = 0.5)
plt.xlabel("Country")
plt.ylabel("Unemployment rate")
plt.title("Five Countries with highest Unemployment rate")
plt.show()
```

10 Countries with highest labor force participation are : -

```python
High_labor = pd.DataFrame(df2.sort_values(by = ['Population: Labor force participation (%)']).tail(5))
High_labor = High_labor.reset_index()
High_labor
```

```python
#Top 5 Countries with highest population density
c = ['red', 'yellow', 'brown', 'blue', 'orange']
plt.bar(High_labor['Country'],High_labor['Population: Labor force participation (%)'],color=c,width = 0.5)
plt.xlabel("Country")
plt.ylabel("Population: Labor force participation (%)")
plt.title("Five Countries with highest Population: Labor force participation (%)")
plt.show()
```

Correlation between Unemployment rate and labor force participation can be seen below : -

```python
corr = df2['Unemployment rate'].corr(df2['Population: Labor force participation (%)'])
```

```python
round(corr, 2)
```

A negative correlation indicates two variables that tend to move in opposite directions. 
In this case as the value of Correlation Coefficient is -0.45, it shows a moderate negative relationship between Unemployment 
rate and Labor force participation.

# 2. Investigate educational enrollment rates and their implications for human capital development.

```python
df3 = df1.groupby('Country')['Gross primary education enrollment (%)','Gross tertiary education enrollment (%)'].sum()
df3
```

	Gross enrollment ratios indicate the capacity of each level of the education system,
    but a high ratio may reflect a substantial number of overage children enrolled in each grade 
    because of repetition or late entry rather than a successful education system. The net enrollment rate 
    excludes overage and underage students and more accurately captures the system's coverage and internal efficiency.

```python
# Top 5 countries with highest Gross primary education enrollment (%)
df3.sort_values(by = ['Gross primary education enrollment (%)']).tail(5)
```

```python
# Top 5 countries with highest Gross tertiary education enrollment (%)
df3.sort_values(by = ['Gross tertiary education enrollment (%)']).tail(5)
```

As the value of Correlation coefficient is almost zero, there is not much relationship between these 2 variales.

# 3. Examine carbon dioxide emissions and their impact on climate change.

```python
df4 = df1.groupby('Country')['Urban_population','Co2-Emissions'].sum()
df4
```

```python
df5 = pd.DataFrame(df4['Co2-Emissions'].nlargest(5))
df5 = df5.reset_index()
df5
```

```python
#PLotting the Graph
plt.figure(figsize = (10,5))
plt.plot(df5.loc[:,'Country'],df5.loc[:,'Co2-Emissions'],color = 'Red',label='Co2 Emissions')
plt.xlabel("Country")
plt.ylabel("Co2 Emissions")
plt.grid(linewidth=0.5)
plt.legend()
```

As we can see from the above data, top 5 countries with the highest Co2 emissions are all the major countries of the world and
are mainly the ones responsible for increasing climate change. Therefore, these countries need to work harder in terms of
climate change and take necessary steps to decrease their Co2 emissions.

```python
corr4 = df4['Co2-Emissions'].corr(df4['Urban_population'])
round(corr4,2)
```

Correlation Coefficient's value is almost 1 shich shows a positive relationship between both the variables.
It means that more the urbanization grows, more will be the Co2 Emissions.

# 5. Analyze population density and land area to study spatial distribution patterns.

```python
df6 = pd.DataFrame(df1.groupby('Country')['Density\n(P/Km2)'].sum())
df6
```

```python
#Top 5 Countries with highest population density
top5_density = df6.sort_values(by = ['Density\n(P/Km2)']).tail(5).reset_index()
c = ['red', 'yellow', 'brown', 'blue', 'orange']
plt.bar(top5_density['Country'],top5_density['Density\n(P/Km2)'],color=c,width = 0.5)
plt.xlabel("Country")
plt.ylabel("Density\n(P/Km2)")
plt.title("Top 5 Countries with highest Population density")
plt.show()
```

```python
#5 Countries with least population density
least5_density = df6.sort_values(by = ['Density\n(P/Km2)']).head(5).reset_index()
c = ['red', 'yellow', 'brown', 'blue', 'orange']
plt.bar(least5_density['Country'],least5_density['Density\n(P/Km2)'],color=c,width = 0.5)
plt.xlabel("Country")
plt.ylabel("Density\n(P/Km2)")
plt.title("Five Countries with least Population Density")
plt.show()
```
